```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # To plot 3D surfaces
import matplotlib
matplotlib.rcParams['text.usetex'] = True  # Allows to use latex in labels
from matplotlib import cm  # colormap
```

```python

# Change the X, Y to a range that includes  $\hat{\beta_0}$, $\hat{\beta_1}$
# remenber np.arange(min, max, step), adjust if necessary
X = np.arange(-5, 5, 0.25)  # X coord of 3D plot corresponds to $\beta_0$ 
Y = np.arange(-5, 5, 0.25)  # Y coord of 3D plot corresponds to $\beta_1$
#######################################
X, Y = np.meshgrid(X, Y)

# change the function for Z to RSS
# Note: you need to use the data from advertising['Radio']
# Hint: Z = sum( ( y_i-beta0 -beta1*x_i )**2 )  summing over the (x_i, y_i) from advertising['Radio']
# remember that the (X, Y) on the 3D plot correspond to (beta0, beta1)
Z = np.sqrt(X**2 + Y**2)  # function to plot 
```

```python
fig = plt.figure()
ax = Axes3D(fig)
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)

# Customize the z axis.
# adjust this if necessary
ax.set_zlim(-1.01, 7.01)

ax.set_xlabel(r'$\beta_0$', fontsize=16)
ax.set_ylabel(r'$\beta_1$', fontsize=16)
ax.set_zlabel('RSS')

plt.show()
```
